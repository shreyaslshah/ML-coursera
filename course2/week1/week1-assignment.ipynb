```python
import turicreate as tc
```

```python
sales = tc.SFrame('./home_data.sframe/')
```

```python
sales
```

```python
train_data,test_data = sales.random_split(.8,seed=0)
```

```python
def simple_linear_regression(input_feature, output):
    sumY = output.sum()
    sumX = input_feature.sum()
    sumXY = 0
    sumX2 = 0
    n = len(input_feature)
    for i in range(0, n):
        sumXY += input_feature[i]*output[i]
        sumX2 += input_feature[i]*input_feature[i]
    slope = (sumXY-(sumY*sumX/n))/(sumX2-(sumX*sumX/n))
    intercept = (sumY/n) - (slope*sumX/n)
    return (intercept, slope)
```

```python
input_feature = train_data['sqft_living']
output = train_data['price']

(squarefeet_intercept, squarfeet_slope) = simple_linear_regression(input_feature, output)

print(squarefeet_intercept, squarfeet_slope)
```

```python
def get_regression_predictions(input_feature, intercept, slope):
    predicted_output = []
    for i in range(0, len(input_feature)):
        predicted_output.append(slope*input_feature[i] + intercept)
    return (predicted_output)
```

```python
print(get_regression_predictions([2650], squarefeet_intercept, squarfeet_slope))
```

```python
def get_residual_sum_of_squares(input_feature, output, intercept, slope):
    RSS = 0
    for i in range(0, len(input_feature)):
        RSS += (pow((output[i] - (slope*input_feature[i] + intercept)), 2))
    return (RSS)
```

```python
print(get_residual_sum_of_squares(
    train_data['sqft_living'], train_data['price'], squarefeet_intercept, squarfeet_slope))
```

```python
def inverse_regression_predictions(output, intercept, slope):
    estimated_input = []
    for i in range(0, len(output)):
        estimated_input.append((output[i] - intercept)/slope)
    return (estimated_input)
```

```python
print(inverse_regression_predictions([800000], squarefeet_intercept, squarfeet_slope))
```
