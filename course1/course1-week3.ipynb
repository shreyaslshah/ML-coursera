```python
products['word_count'] = tc.text_analytics.count_words(products['review'])
products.head()
```

## expolore a product (vulli sophie)

```python
giraffe_reviews = products[products['name'] == 'Vulli Sophie the Giraffe Teether']
len(giraffe_reviews)
```

```python
giraffe_reviews['rating'].show()
```

# define positive and negative sentiment

```python
# ignore all 3 star review because they are neutral
products = products[products['rating']!=3]

# sentiment will be 1 for positive and 0 for negative
products['sentiment'] = products['rating'] >= 4

products.head()
```

```python
training_data, test_data = products.random_split(0.8, seed = 0)
```

```python
sentiment_model = tc.logistic_classifier.create(training_data, 
                                                target = 'sentiment', 
                                                features = ['word_count'], 
                                                validation_set = test_data)
```

```python
sentiment_model.evaluate(test_data, metric='roc_curve')
sentiment_model.evaluate(test_data)
```

# apply the model

```python
giraffe_reviews['predicted_sentiment'] = sentiment_model.predict(giraffe_reviews,
                                                                 output_type='probability')
giraffe_reviews.head()
```

## sort the reviews based on the predicted sentiment and explore

```python
giraffe_reviews = giraffe_reviews.sort('predicted_sentiment', ascending=False)
giraffe_reviews.head()
```

# assignment


```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
i=0

def modify_count(dic):
  if selected_words[i] in dic:
    return dic[selected_words[i]]
  else:
    return 0

for word in selected_words:
  products[word] = products['word_count'].apply(modify_count)
  i+=1


products.head()
```

```python
import sys
mini = sys.maxsize
maxi = -sys.maxsize
minword = ''
maxword = ''

for word in selected_words:
  sum = products[word].sum()
  if mini > sum:
    mini = sum
    minword = word
  if maxi < sum:
    maxi = sum
    maxword = word

print(minword, maxword)
```

```python
training_data,test_data = products.random_split(.8, seed=0)

selected_words_model = tc.logistic_classifier.create(training_data, 
                                                target = 'sentiment', 
                                                features = selected_words, 
                                                validation_set = test_data)
```

```python
selected_words_model.coefficients.sort('value', ascending=False)
```

```python
selected_words_model.evaluate(test_data)
```
